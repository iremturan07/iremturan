```python
import tkinter as tk
from tkinter import messagebox

class ToDoListApp:
    def __init__(self, root):
        self.root = root
        self.root.title("To-Do List Uygulaması")

        self.tasks = []

        self.task_listbox = tk.Listbox(self.root, height=10, width=50, borderwidth=2, relief="groove")
        self.task_listbox.pack(pady=10)

        self.task_entry = tk.Entry(self.root, width=50)
        self.task_entry.pack()

        btn_frame = tk.Frame(self.root)
        btn_frame.pack(pady=20)

        add_btn = tk.Button(btn_frame, text="Görev Ekle", command=self.add_task)
        add_btn.grid(row=0, column=0, padx=10)

        del_btn = tk.Button(btn_frame, text="Görevi Sil", command=self.delete_task)
        del_btn.grid(row=0, column=1, padx=10)

        clear_btn = tk.Button(btn_frame, text="Listeyi Temizle", command=self.clear_tasks)
        clear_btn.grid(row=0, column=2, padx=10)

        self.display_tasks()

    def add_task(self):
        task = self.task_entry.get()
        if task:
            self.tasks.append(task)
            self.task_listbox.insert(tk.END, task)
            self.task_entry.delete(0, tk.END)
        else:
            messagebox.showwarning("Uyarı", "Lütfen bir görev girin.")

    def delete_task(self):
        try:
            index = self.task_listbox.curselection()[0]
            self.task_listbox.delete(index)
            self.tasks.pop(index)
        except IndexError:
            messagebox.showwarning("Uyarı", "Lütfen silinecek bir görev seçin.")

    def clear_tasks(self):
        self.task_listbox.delete(0, tk.END)
        self.tasks.clear()

    def display_tasks(self):
        for task in self.tasks:
            self.task_listbox.insert(tk.END, task)


if __name__ == "__main__":
    root = tk.Tk()
    app = ToDoListApp(root)
    root.mainloop()
```
